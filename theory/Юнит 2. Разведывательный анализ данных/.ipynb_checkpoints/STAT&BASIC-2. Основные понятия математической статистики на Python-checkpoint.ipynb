# 2.3. Распределение вероятностей дискретной случайной величины
## Распределение вероятностей дискретной случайной величины
Распределение вероятностей дискретной случайной величины — это набор ее возможных значений и их соответствующих вероятностей. Распределением этот набор называется потому, что он показывает, каким образом общая вероятность в 100% распределена по возможным исходам, какие исходы более или менее вероятны.

Для дискретных случайных величин распределение вероятностей, как правило, задается таблицей или графиком.
