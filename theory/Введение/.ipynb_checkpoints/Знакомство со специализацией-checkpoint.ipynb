# Что такое Data Science?

Data Science — это наука о работе с большими данными: их сборе и обработке, анализе и выявлении взаимосвязей с помощью машинного обучения или нейронных сетей. На основе найденных взаимосвязей затем строятся прогнозные модели, предсказывающие целевой показатель. Инсайты, полученные в результате работы дата-сайентиста, затем применяются в самых разнообразных областях: бизнесе, медицине, промышленности, науке и так далее.

Сферы применения Data Science поистине безграничны: оптимизация энергопотребления на заводах и целых городах, определение риска заболеваний, предсказание вероятности дефолта компаний, прогнозы цен на товары, рекомендательные системы, машинный перевод, голосовые помощники, беспилотники… Хотите в этом убедиться? Прочитайте этот длинный список Data Science-решений для разных областей.

Иногда технологии Data Science находят необычное применение. Вот пара примеров из сферы разработки искуcственного интеллекта (подраздел Data Science).

ПРИМЕР 1  
Всем нам знакомы чат-боты, которые помогают пользователям компаний с тем или иным вопросом. Инженеры Facebook пошли дальше и решили создать бота-переговорщика: его задача была договориться о сделке на самых выгодных условиях. После тренировки на 5,8 тысячах настоящих диалогов, бот научился… блефовать, умело скрывая свой интерес к товару до подходящего момента.

Результаты первого конкурса были настолько впечатляющими, что мероприятие решили проводить каждый год. Можно выделить несколько типов задач, которые решают специалисты Data Science:

→ КЛАССИФИКАЦИЯ	Задача заключается в распределении объектов по изначально заданным признакам. К примеру, фотографии собак модель отнесет к группе «животные», фотографии цветов — к группе «растения» и так далее. Часто используется в спам-фильтрах, поиске похожих объектов, распознавании рукописного текста и др.
→ РЕГРЕССИЯ	В таких типах задач нужно найти зависимость одного параметра от другого и предсказать его значение. Пример — предсказания биржевых котировок, погоды или пробок на дорогах.
→ КЛАСТЕРИЗАЦИЯ	При решении задач кластеризации алгоритм группирует объекты по похожим признакам. Это используется при распознавании объектов на фото, в поисковых механизмах, социологических исследованиях.
→ ОБНАРУЖЕНИЕ АНОМАЛИЙ	ИИ ищет данные, которые по тем или иным признакам выбиваются из общего массива. Наиболее часто это используется в банковской сфере для обнаружения мошенников.
→ АССОЦИАЦИЯ	Задача заключается в нахождении закономерностей между тем как признаки одного объекта соотносятся с признаками других объектов. Пример: если вы заказываете в интернет-магазине пряжу, сайт может порекомендовать вам наборы спиц и книги со схемами для вязания, потому что все эти предметы имеют общий признак: они относятся к вязанию.
С Data Science тесно связаны понятия ИИ, нейронной сети и машинного обучения. С ними мы познакомимся дальше в курсе.

# Поток работ (Workflow) Data Science
Для стороннего наблюдателя работа дата-сайентиста выглядит примерно так: мы берём данные, потом с ними что-то происходит в чёрном ящике, после чего мы получаем предсказания.



Так как вы пришли сюда именно для того, чтобы не быть сторонними наблюдателями, давайте разберёмся, что же происходит внутри нашего чёрного ящика.

За любыми эффективными процессами стоит проверенная временем методология. Наиболее популярной и распространенной методологией ведения DS-проектов до сих пор является Cross-Industry Standard Process for Data Mining, или сокращенно CRISP-DM, которая была разработана ещё в 1996 году.

ЭТАП 1
Анализ бизнес-целей и определение ключевых бизнес-метрик

На этом этапе вы должны ответить на вопросы:

Какую задачу бизнеса должна решить ваша модель?
Какими метриками мы измерим полезность от внедрения модели? Это должна быть именно бизнес-метрика, например конверсия в покупку, ретеншн (показатель удержания), различные финансовые показатели. 
На основе полученной от продакт-менеджера информации вы сможете определиться с типом задачи, с тем, какие данные вам понадобятся для её решения, правильно выбрать технические метрики для будущей модели. Главное — вы будете знать, как донести результаты вашей работы для продакт-менеджера, который может плохо разбираться в Data Science. 

На этом этапе дата-сайентисту важно: 

понимать особенности той сферы, в которой он трудится. Всё-таки задачи, которые решают специалисты в ритейле, сильно отличаются от задач в сфере страхования или финансов. 
переводить бизнес-задачу на язык Data Science. Уже на этом этапе вы должны понимать, решается ли задача вообще, и если да, то какой объём работы предстоит выполнить, какой алгоритм вы будете использовать и какие данные будут нужны.
ПОМНИТЕ: МОДЕЛЬ В ВАКУУМЕ НИКОМУ НЕ НУЖНА!

После того, как вы составили ТЗ, можно приступать к сбору пайплайна.

Пайплайн (от англ. pipeline — трубопровод) — это путь, который проходят данные от сбора и анализа до работы модели машинного обучения в реальной среде. Качество пайплайна определяется тем, насколько он автоматизирован, гибок и при этом устойчив.

ЭТАП 2
Сбор и хранение данных

Чтобы обучаться, модели нужны данные. И чем сложнее модель, чем больше параметров она будет анализировать, тем больше нужно данных. 

Машинное обучение делится на два типа: обучение с учителем (Supervised Learning) и без (Unsupervised Learning). 

Обучение с учителем подразумевает, что мы даём модели заранее размеченные данные с правильным ответом (например, какие товары чаще покупаются вместе, данные о поведении пользователей на сайте и их конверсия в покупку и так далее) 

Как вы думаете, если мы будем обучать модель на некачественных данных, каков будет результат её работы? Конечно, нулевой. Как гласит ещё одно правило: Garbage in — garbage out (сокращенно GIGO, «мусор на входе — мусор на выходе»).

Существуют различные способы собрать нужные вам данные:

Вытащить их из базы данных продукта.
Использовать общедоступные датасеты (например, на сайтах организаций, занимающихся статистическими исследованиями, или на github).
Спарсить (собрать) данные из интернета.
Для некоторых особо сложных задач вам могут понадобиться услуги компаний, занимающихся ручной разметкой данных (либо делать это самостоятельно…).
Если вы когда-либо проходили проверку Google reCaptcha, в которой нужно найти на изображении автомобиль, светофор или витрину, знайте, что вы приняли участие в сборе данных для обучения нейросети, которую Google планирует использовать для управления беспилотными автомобилями. Решая такую несложную для человека задачку, пользователи, сами того не зная, помогают откалибровать и проверить работу искусственного интеллекта.



На этом этапе дата-сайентисту пригодятся умения:

- собирать данные из файлов разных форматов;
- получать данные из веб-источников (парсинг сайтов и подключение через API);
- работать с хранилищами данных разных типов (MySQL, PostgresSQL, MongoDB);
- уметь писать SQL-запросы к реляционной БД;
- работать с распределенными СУБД (Hadoop, Apache Spark).

ЭТАП 3
Обработка и очистка сырых данных

Это, пожалуй, самый важный этап работы дата-сайентиста: как вы помните, именно качество данных на 80 % определяет результат работы модели. 

На этом этапе вы анализируете полученные данные и выявляете ошибки, пропущенные значения, дубликаты, некорректные форматы, определяете признаки, с которыми будете работать. А затем приводите свой датасет к нужному виду. 

Требования к специалисту:

- уметь работать с датафреймами в Python;
- знать основные методы предобработки данных: находить выбросы, заменять пропущенные значения, преобразовывать данные в нужный тип переменных (Pandas).

ЭТАП 4
Разведывательный анализ данных (EDA)

На этом этапе дата-сайентист превращается в настоящего детектива и исследует данные на наличие интересных взаимосвязей и выбирает значимые переменные для построения будущей модели. Как и в любом исследовании, важно в самом начале правильно задать вопрос — сформулировать гипотезу, которую вы будете проверять. 

Так как на этом шаге вы будете проводить статистический анализ и строить много графиков и диаграмм, то вам помогут следующие умения:

- применять библиотеку NumPy для статистического анализа;
- строить матрицы корреляций и выявлять скрытые зависимости;
- визуализировать данные с помощью Pandas, Matplotlib и Seaborn;
- обогащать данные новыми переменными на основе имеющихся (самое время проявить креативность!) — Feature Engineering.

ЭТАП 5
Построение модели

Момент, вознаграждающий нас за все предыдущие труды! Мы наконец-то понимаем, зачем до этого часами готовили и анализировали данные. 



Как проходит процесс моделирования? Вы тренируете модель на подготовленных данных, а затем оцениваете результат: каково значение целевой метрики? Где алгоритм допустил ошибку? Как можно улучшить результат? 

Совет от практиков. Двигайтесь итерациями. Это значит, что вначале достаточно применить самый базовый алгоритм и на его основе построить baseline (базовое решение). Что нужно уметь делать на этом этапе?

- выбрать и и применить подходящие для задачи алгоритмы ML или DL (Deep Learning);
- владеть библиотеками ML и DL;
- оценить качество работы модели и оптимизировать её работу;
- понимать внутреннюю работу алгоритмов, а значит разбираться в математике, лежащей в их основе.

ЭТАП 6
Оценка результативности

После того, как вы добились целевых показателей технических метрик, пришло время презентовать результаты продакт-менеджеру. Вспоминаем те бизнес-задачи, которые наша модель должна решать, и пишем отчёт. 

Именно то, как ваша модель решает поставленную бизнес-задачу, повлияет на последующее решение — выводить ли её в продакшн. К тому же бывают случаи, когда технические метрики просто не могут полностью отразить работу алгоритма в реальной среде, и тогда комментарии продакт-менеджера будут очень полезны. 

Ключевое умение на этом этапе: навык презентации.

ЭТАП 7
Вывод модели в продакшн

Если защита проекта прошла удачно, то команда примет решение выводить модель в продакшн (этот процесс ещё называют **деплоем модели**).

Что это вообще значит? Это значит, что ваша модель будет выведена на основной сервер и начнет влиять на работу продукта. 

В небольших компаниях деплоем моделей занимается всё тот же дата-сайентист (человек-оркестр!). Однако чем сложнее продукт и чем выше цена ошибки, тем скорее вам потребуется помощь дата-инженера.

Однако даже в случае, когда деплоем модели будет заниматься другой человек, дата-сайентисту важно понимать, как модель будет использована в продакшне, знать требования продакшн-среды и уметь протестировать модель до её деплоя. Ну и общаться с коллегой на одном языке тоже никогда не бывает лишним!

Ключевые навыки на этом шаге:

- понимать, как работают сервисы и уметь написать свой, хотя бы даже самый простой (Flask, HTTP);
- как настроить мониторинг работы модели (обновление метрик, ведение логов ошибок);
- как контейнеризовать приложения (Docker и Docker Compose).
