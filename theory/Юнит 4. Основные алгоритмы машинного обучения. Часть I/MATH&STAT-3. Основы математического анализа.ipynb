# 3.2. Что такое функция?
Определение: Функцией в математике называется правило, по которому можно сопоставить элементы двух множеств.

Формальное определение: Для любого элемента х из множества Х существует единственный элемент y из множества Y, такой что f(x) = y:


Проиллюстрировав на картинке множество Х и множество Y, видим: у каждого элемента x из X должна быть пара y в Y. Таким образом, под действием функции f все множество X переходит в некоторую часть множества Y.


Восклицательный знак в формальном определении функции говорит о том, что такая пара x – y должна быть одна. Это значит, что нельзя одному х поставить в соответствие два разных y. И в то же время двум х — один y: сколько угодно!

Для создания последовательностей чисел в numpy имеется функция numpy.arange()(https://numpy.org/doc/stable/reference/generated/numpy.arange.html), аналогичная встроенной в python range(), только вместо списков она возвращает массивы и принимает не только целые значения.

```python
#  импортируем необходимые библиотеки для работы с массивами чисел и для визуализации
import numpy as np
import matplotlib.pyplot as plt

# х - аргумент в нашей функции. 
# передадим начало и конец числового отрезка нашей функции и шаг аргумента:
x = np.arange(-10, 10.01, 0.01) # массив значений аргумента

# построение графика параболы
plt.plot(x, x**2) 

# включим сетку на графике 
plt.grid(True)

# показать график
plt.show() 
```

Линеаризация
Определение: Задача о приближении, с помощью которой мы только что ввели понятие производной, называется линеаризацией.

Чем ближе новая точка расположена к старой, тем точнее приближение.

На идее линеаризации основываются marginal effects в нелинейных регрессиях.
